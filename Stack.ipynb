```python
class StackNode:
    def __init__ (self, value):
        self.value = value;
        self.next = None;
        
class Stack:
    def __init__(self):
        self.size = 0;
        self.top = None;
        
    def __len__(self):
        return self.size;
    
    def isEmpty(self):
        return self.size == 0;
    
    def pop(self):
        assert not (self.isEmpty());
        node = self.top;
        self.size -= 1;
        self.top = self.top.next;
        return node.value;
    
    def push(self, value):
        newNode = StackNode(value);
        newNode.next = self.top;
        self.top = newNode;
        self.size += 1;
        
        
    def __str__(self):
        if self.isEmpty():
            return "[]"
        s = "[";
        currentnode = self.top;
        while (currentnode is not None):
            if currentnode.next is not None:
                s += str(currentnode.value) + ", ";
            else:
                s += str(currentnode.value) + "]"
            currentnode = currentnode.next;
        return s;
    
    def peek(self):
        assert not self.isEmpty();
        return self.top.value;
```

```python
import math;

class PostfixCalculator:
    def __init__(self):
        self.opStack = Stack();
        self.saveStack = Stack();
        
    def value(self, x):
        self.opStack.push(x);
        
    def result(self):
        if self.opStack.isEmpty():
            return None;
        return self.opStack.peek();
    
    def clear(self):
        while (not self.opStack.isEmpty()):
            self.opStack.pop();
            
    def clearLast(self):
        if not self.opStack.isEmpty():
            self.opStack.pop();
            
    def compute(self, op):
        assert len(self.opStack) > 0;
        if op == "+":
            assert len(self.opStack) > 1;
            a = self.opStack.pop();
            b = self.opStack.pop();
            self.opStack.push(b + a);
        elif op == "-":
            assert len(self.opStack) > 1;
            a = self.opStack.pop();
            b = self.opStack.pop();
            self.opStack.push(b - a);
        elif op == "/":
            assert len(self.opStack) > 1;
            a = self.opStack.pop();
            b = self.opStack.pop();
            self.opStack.push(b / a);
        elif op == "*":
            assert len(self.opStack) > 1;
            a = self.opStack.pop();
            b = self.opStack.pop();
            self.opStack.push(b * a);
        elif op == "**":
            assert len(self.opStack) > 1;
            a = self.opStack.pop();
            b = self.opStack.pop();
            self.opStack.push(b ** a);
        elif op == "abs":
            assert len(self.opStack) > 0;
            value = self.opStack.pop();
            self.opStack.push(abs(value));
        elif op == "sqrt":
            assert len(self.opStack) > 0;
            value = self.opStack.pop();
            self.opStack.push((value**(1/2)));
        elif op == "sin":
            assert len(self.opStack) > 0;
            value = self.opStack.pop();
            self.opStack.push(math.sin(value));
        elif op == "cos":
            assert len(self.opStack) > 0;
            value = self.opStack.pop();
            self.opStack.push(math.cos(value));
        elif op == "tan":
            assert len(self.opStack) > 0;
            value = self.opStack.pop();
            self.opStack.push(math.tan(value));
        
    def store(self):
        assert len(self.opStack) > 0;
        self.saveStack.push(self.opStack.pop());
                         
    def recall(self):
        assert len(self.saveStack) > 0;
        self.opStack.push(self.saveStack.pop());
                         
calculator = PostfixCalculator();
calculator.value(5);
calculator.value(10);
calculator.compute("+")
print("5 + 10: ", calculator.result());
calculator.value(12);
calculator.value(1);
calculator.compute("*")
print("12 * 1: ", calculator.result());
calculator.compute("-")
print("15 - 12: ", calculator.result());
calculator.value(12);
calculator.value(2);
calculator.compute("/")
print("12 / 2: ", calculator.result());
calculator.compute("**");
print("3 ** 6: ", calculator.result());
calculator.compute("cos");
print("cos(729): ", calculator.result());
calculator.value(10);
calculator.compute("*");
print("0.98 * 10: ", calculator.result());
calculator.compute("tan");
print("tan(9.88): ", calculator.result());
calculator.compute("sin");
print("sin(0.49): ", calculator.result());
calculator.value(10);
calculator.compute("-");
print("0.47 - 10: ", calculator.result());
calculator.compute("abs");
print("abs(-9.52): ", calculator.result());
```
