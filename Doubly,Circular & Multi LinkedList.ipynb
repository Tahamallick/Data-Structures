```python
class Node:
    def __init__(self, value):
        self.value = value;
        self.next = None;
        self.prev = None
        
class DoublyLinkedList:
    def __init__(self):
        self.head = None;
        self.count = 0;
        
    def __len__(self):
        return self.count;
    
    def insertAtStart(self, value):
        newnode = Node(value);
        if self.head is not None:
            newnode.next = self.head;
            self.head.prev = newnode;
            self.head = newnode;
        else:
            self.head = newnode;
        self.count += 1; 
        
    def insertAtEnd(self, value):
        newnode = Node(value);
        currentnode = self.head;
        if currentnode is not None:
            while (currentnode.next is not None):
                currentnode = currentnode.next;
            currentnode.next = newnode;
            newnode.prev = currentnode;
        else:
            self.head = newnode;
        self.count += 1;
        
    
    def search(self, value):
        if len(self) > 0:
            currentnode = self.head;
            while (currentnode.next is not None and value != currentnode.value):
                currentnode = currentnode.next;
            if currentnode.value == value:
                return currentnode, True
        return currentnode, False;
    
    def delete(self, value):
        searchParam = self.search(value);
        assert searchParam[1];
        currentnode = searchParam[0];
        if currentnode.prev == None and currentnode.next == None:
            self.head = None;
        elif currentnode.prev == None:
            currentnode.next.prev = None;
            self.head = currentnode.next;
        elif currentnode.next == None:
            currentnode.prev.next = None;
        else:
            currentnode.prev.next = currentnode.next;
            currentnode.next.prev = currentnode.prev;
        self.count -= 1;
        
    def __str__(self):
        s = "[";
        currentnode = self.head;
        while (currentnode is not None):
            s += str(currentnode.value) + ", ";
            currentnode = currentnode.next;
        s += "]";
        return s;
    
d = DoublyLinkedList();
d.insertAtStart(23);
d.insertAtEnd(2323);
print(d);
d.insertAtStart(14);
d.insertAtEnd(263);
print(d);
d.delete(14);
print(d);
d.delete(2323);
print(d);
d.delete(263);
print(d);
d.delete(23);
print(d);
```

```python
class DoublyNode:
    def __init__ (self, value):
        self.value = value;
        self.next = None;
        self.prev = None;
        
class DoublyCircularLinkedListQueue:
    def __init__(self):
        self.head = None;
        self.tail = None;
        self.count = 0;
        
    def __len__(self):
        return self.count;
    
    def first(self):
        assert len(self) > 0;
        return self.head;
    
    def enqueue(self, value):
        newnode = DoublyNode(value);
        currentnode = self.head;
        if currentnode is not None:
            while(currentnode.next is not self.head):
                currentnode = currentnode.next;
            currentnode.next = newnode;
            newnode.prev = currentnode
            newnode.next = self.head;
            self.tail = newnode;
            self.tail.next = self.head;
            self.head.prev = self.tail;
        else:
            self.tail = newnode;
            self.head = newnode;
            self.tail.next = self.head;
            self.tail.prev = self.head;
            self.head.next = self.tail;
            self.head.prev = self.tail;
        self.count += 1;
        
    def dequeue(self):
        assert len(self) > 0;
        if len(self) != 1:
            node = self.head;
            self.head = self.head.next;
            self.head.prev = self.tail;
            self.tail.next = self.head;
        else:
            node = self.head;
            self.tail = None;
            self.head = None;
        self.count -= 1;
        return node;
        
    
    def rotate(self):
        self.tail = self.tail.next;
        self.dequeue()
        
        
    def __str__(self):
        s = "[";
        currentnode = self.head;
        if len(self) > 0:
            while (currentnode.next is not self.head):
                s += str(currentnode.value) + ", ";
                currentnode = currentnode.next;
            s += str(currentnode.value) + "]";
        else:
            return "[]";
        return s;
        
        
p = DoublyCircularLinkedListQueue();
p.enqueue(90);
p.enqueue(912);
p.enqueue(3240);
p.enqueue(2);
print(p)
p.dequeue()
print(p);
p.dequeue()
print(p);
p.dequeue()
print(p)
p.dequeue()
print(p)
p.enqueue(90);
p.enqueue(912);
p.enqueue(3240);
p.enqueue(2);
print(p)
p.rotate()
print(p)
p.dequeue();
print(p)
```
