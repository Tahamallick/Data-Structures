```python
class Node:
    def __init__(self, value):
        self.value = value;
        self.next = None;
        
class QueueLinkedList:
    def __init__(self):
        self.tail = None;
        self.head = None;
        self.count = 0;
        
    def __len__(self):
        return self.count;
    
    def enqueue(self, value):
        newnode = Node(value);
        if len(self) == 0:
            self.head = newnode;
            self.tail = self.head;
        else:
            self.head.next = newnode;
            self.head = newnode;
        self.count += 1;
        
    def dequeue(self):
        assert len(self) > 0;
        node = self.tail;
        self.tail = self.tail.next;
        self.count -= 1;
        return node;
    
    def __str__(self):
        if len(self) > 0:
            s = "";
            currentnode = self.tail;
            while(currentnode is not None):
                s += str(currentnode.value) + ", ";
                currentnode = currentnode.next;
            s = s[:len(s) -2];
            return s;
        return "None";
    
q = QueueLinkedList();
q.enqueue(12);
print(q);
q.enqueue(32);
print(q);
q.enqueue(142);
print(q);
q.enqueue(1345);
print(q);
q.enqueue(-43);
print(q);
q.dequeue();
print(q);
q.dequeue();
print(q);
q.dequeue();
print(q);
q.dequeue();
print(q);
q.dequeue();
print(q);
```

```python
class Task:
    def __init__(self, value, priority):
        self.value = value;
        self.priority = priority;
        
class PriorityQueue:
    def __init__(self):
        self.list = [];
    
    
    def __len__(self):
        return len(self.list);
    
    def enqueue(self, value, priority):
        task = Task(value, priority);
        self.list.append(task);
        
    def dequeue(self):
        assert len(self) > 0;
        index = 0;
        highestPriority = self.list[index].priority;
        for i in range(len(self)):
            if self.list[i].priority < highestPriority:
                highestPriority = self.list[i].priority;
                index = i;
        return self.list.pop(index);
    
    def __str__(self):
        s = "[ ";
        for i in self.list:
            s += "[" + str(i.value) + ", " + str(i.priority) +"] ";
        s += "]";
        return s;
    
a = PriorityQueue();
a.enqueue(5, 1);
a.enqueue(10, 2);
a.enqueue(12, 4);
a.enqueue(434, 1);
a.enqueue(213425, 2);
print(a);
a.dequeue();
print(a);
a.dequeue();
print(a);
a.dequeue();
print(a);
a.dequeue();
print(a);
a.dequeue();
print(a);
```
